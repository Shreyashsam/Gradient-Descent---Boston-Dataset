```python
def step_gradient(points, learning_rate, m):
    m_slope = np.zeros(14)
    M = len(points)
    for i in range(M):
        x = points[i]
        y = Y[i]
        for j in range(14):
            m_slope[j] = m_slope[j]+(-2/M)* (y - (m * x).sum() )*x[j]
    for j in range(14):
        m[j] = m[j] - learning_rate*m_slope[j]
    return m
```

```python
def gd(points, learning_rate, num_iterations):
    m=np.zeros(14)
    for i in range(num_iterations):
        m = step_gradient(points, learning_rate, m)
        print(i, " Cost: ", cost(points, m))
    return m
```

```python
def cost(points, m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i]
        y = Y[i]
        total_cost += (1/M)*((y - (m*x).sum() )**2)
    return total_cost
```

```python
def run():
    learning_rate = 0.1
    num_iterations = 1500
    m = gd(points, learning_rate, num_iterations)
    print(m)
    return m
```

```python
m=run()
```
